```python
import numpy as np
import pandas as pd
import re
```

## import data into df

```python

df=pd.read_csv('DS_DA_BS.csv')
df=df[['Salary Estimate','Location','Company Name','Sector','role']]
```

clean up the salary data

```python
for i in df.index:
    if re.search("Hour|-1",df['Salary Estimate'][i]):
        df.drop([i],inplace=True)
df.shape
```

```python
df.replace(['-1'], [np.nan], inplace=True)
df.replace(['-1.0'], [np.nan], inplace=True)
df.replace([-1], [np.nan], inplace=True)
```

```python
def extract_numbers_max(x):
    temp=re.findall(r'[0-9]+',x)
    return max([int(i) for i in temp])

def extract_numbers_min(x):
    temp=re.findall(r'[0-9]+',x)
    return min([int(i) for i in temp])

df['min salary']=df['Salary Estimate'].apply(extract_numbers_min)
df['max salary']=df['Salary Estimate'].apply(extract_numbers_max)
df['mean salary']=(df['min salary']+df['max salary'])/2
```

clean up location column

```python
def state_extract(x):
    return x.split(',')[-1]
df['states']=df['Location'].apply(state_extract)
df
```

```python
# df['states'].unique()
df['states'].value_counts()
```

```python
selected_states=[' TX',' CA',' NY',' IL',' AZ',' PA',' FL',' OH',' NJ']
df=df[df['states'].isin(selected_states)]
df
```

### clean up sector column

```python
df['Sector'].value_counts().head()
# df['Sector'].unique()
# df.shape
```

```python
selected_sector=['Consumer Services',
       'Information Technology', 
                 'Business Services', 'Insurance',
       'Finance', 'Retail', 'Media', 
        'Non-Profit', 'Education', 'Government',
       'Health Care', 'Oil, Gas, Energy & Utilities',
       'Accounting & Legal', 'Manufacturing', 'Real Estate',
       'Biotech & Pharmaceuticals', 
       'Aerospace & Defense', 'Construction, Repair & Maintenance',
       'Transportation & Logistics', 'Telecommunications']
df=df[df['Sector'].isin(selected_sector)]
df['Sector'].value_counts()
```

## begin coding here

```python
data=df[['role','Sector','states','min salary','max salary','mean salary']]
data.to_csv('cleaned_data.csv')
```

```python
data=pd.read_csv('cleaned_data.csv')
data[(data['role']=='DS') & (data['Sector']=='Information Technology')]['states'].value_counts()
```

```python
df_selection=data[data['Sector'].isin(['Insurance','Retail'])]
states=[' NY',' CA',' TX',' PA']
roles=['DS','BA','DA']

df_states_count=pd.DataFrame(
        np.zeros((len(states),len(roles))),
        index=states,
        columns=roles)

for i in states:
    for j in roles:
        df_states_count.at[i,j]=len(
            df_selection[(df_selection['states']==i) & (df_selection['role']==j)]
        )
df_states_count
```
