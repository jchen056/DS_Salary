```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
```

## import data into df

```python

df=pd.read_csv('DS_DA_BS.csv')
df=df[['Salary Estimate','Location','Company Name','Sector','role','Revenue','Size']]
df['Size'].unique()
```

clean up the salary data

```python
for i in df.index:
    if re.search("Hour|-1",df['Salary Estimate'][i]):
        df.drop([i],inplace=True)
df.shape
```

```python
df.replace(['-1'], [np.nan], inplace=True)
df.replace(['-1.0'], [np.nan], inplace=True)
df.replace([-1], [np.nan], inplace=True)
```

```python
def extract_numbers_max(x):
    temp=re.findall(r'[0-9]+',x)
    return max([int(i) for i in temp])

def extract_numbers_min(x):
    temp=re.findall(r'[0-9]+',x)
    return min([int(i) for i in temp])

df['min salary']=df['Salary Estimate'].apply(extract_numbers_min)
df['max salary']=df['Salary Estimate'].apply(extract_numbers_max)
df['mean salary']=(df['min salary']+df['max salary'])/2
```

clean up location column

```python
def state_extract(x):
    return x.split(',')[-1]
df['states']=df['Location'].apply(state_extract)
df.isna().sum()
```

```python
# df['states'].unique()
df['states'].value_counts()
```

```python
selected_states=[' TX',' CA',' NY',' IL',' AZ',' PA',' FL',' OH',' NJ']
df=df[df['states'].isin(selected_states)]
df
```

### clean up sector column

```python
df['Sector'].value_counts().head()
# df['Sector'].unique()
# df.shape
```

```python
selected_sector=['Consumer Services',
       'Information Technology', 
                 'Business Services', 'Insurance',
       'Finance', 'Retail', 'Media', 
        'Non-Profit', 'Education', 'Government',
       'Health Care', 'Oil, Gas, Energy & Utilities',
       'Accounting & Legal', 'Manufacturing', 'Real Estate',
       'Biotech & Pharmaceuticals', 
       'Aerospace & Defense', 'Construction, Repair & Maintenance',
       'Transportation & Logistics', 'Telecommunications']
df=df[df['Sector'].isin(selected_sector)]
df['Sector'].value_counts()
df.shape
```

clean up revenue column

```python
df['Revenue'] = df['Revenue'].replace('Unknown / Non-Applicable', None)
set(df['Revenue'])
```

```python
df['Revenue'] = df['Revenue'].str.replace('$', ' ')
df['Revenue'] = df['Revenue'].str.replace('(USD)', ' ')
df['Revenue'] = df['Revenue'].str.replace('(', ' ')
df['Revenue'] = df['Revenue'].str.replace(')', ' ')
df['Revenue'] = df['Revenue'].str.replace(' ', '')
set(df['Revenue'])
```

```python
df['Revenue'] = df['Revenue'].str.replace('2to5billion','2billionto5billion')
df['Revenue'] = df['Revenue'].str.replace('1to2billion','1billionto2billion')
df['Revenue'] = df['Revenue'].str.replace('5to10billion','5billionto10billion')
df['Revenue'] = df['Revenue'].replace('10+billion', '10billionto11billion')
set(df['Revenue'])
```

```python
df['Revenue'] = df['Revenue'].str.replace('Lessthan1million', '0millionto1million')
df['Revenue'] = df['Revenue'].str.replace('million', ' ')
df['Revenue'] = df['Revenue'].str.replace('billion', '000 ')
df['Revenue'].value_counts()

df=df[df['Revenue'].isin(['100to500 ', '500 to1000 ', '10000 to11000 ', '25to50 ',
       '5000 to10000 ', '5to10 ', '50to100 ', '1000 to2000 ',
       '2000 to5000 ', '0 to1 ', '1to5 ', '10to25 '])]
df
```

```python
Revenue = df['Revenue'].str.split("to",expand=True)
df['Minimum Revenue'] = Revenue[0]
df['Maximum Revenue'] = Revenue[1]
df['Maximum Revenue'] = pd.to_numeric(df['Maximum Revenue'])
df['Minimum Revenue'] = pd.to_numeric(df['Minimum Revenue'])
len(df)
```

```python
df.dropna(inplace=True)
df=df.reset_index(drop=True)
df
df.groupby(['Revenue'])['role'].aggregate(['count']).sort_values(by=['count'],ascending=False)
```

## begin coding here

```python
data=df[['role','Sector','states','min salary','max salary','mean salary']]
data.to_csv('cleaned_data.csv')
```

```python
data=pd.read_csv('cleaned_data.csv')
data[(data['role']=='DS') & (data['Sector']=='Information Technology')]['states'].value_counts()
```

```python
df_selection=data[data['Sector'].isin(['Insurance','Retail'])]
states=[' NY',' CA',' TX',' PA']
roles=['DS','BA','DA']

df_states_count=pd.DataFrame(
        np.zeros((len(states),len(roles))),
        index=states,
        columns=roles)

for i in states:
    for j in roles:
        df_states_count.at[i,j]=len(
            df_selection[(df_selection['states']==i) & (df_selection['role']==j)]
        )
df_states_count
```
